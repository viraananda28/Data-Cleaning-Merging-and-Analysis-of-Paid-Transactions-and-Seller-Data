1. Membuat kolom baru bernama "Full Name"

```python
import pandas as pd
```

```python
# Membaca file Excel
excel_file = "Homework-DataFrame-Dataset.xlsx"
```

```python
# Membaca data dari sheet "Paid-Transaction" dan "Seller"
paid_transaction_df = pd.read_excel(excel_file, sheet_name="Paid-Transaction", header=1)
seller_df = pd.read_excel(excel_file, sheet_name="Seller", header=1)
```

```python
paid_transaction_df.head()
```

```python
paid_transaction_df = paid_transaction_df.iloc[1:,1:]
paid_transaction_df.reset_index(drop=True, inplace = True)
paid_transaction_df.head()
```

```python
paid_transaction_df.dtypes
```

```python
paid_transaction_df['Paid Date'] = pd.to_datetime(paid_transaction_df['Paid Date'], format='%Y%m%d', errors='ignore')
paid_transaction_df.head()
```

```python
paid_transaction_df.dtypes
```

```python
seller_df.head()
```

```python
seller_df = seller_df.iloc[1:,1:]
seller_df.reset_index(drop=True, inplace = True)
seller_df.head()
```

```python
# Menampilkan nama kolom di DataFrame "Paid-Transaction"
print("Columns in 'Paid-Transaction' DataFrame:", paid_transaction_df.columns)
```

```python
# Menampilkan nama kolom di DataFrame "Seller"
print("Columns in 'Seller' DataFrame:", seller_df.columns)
```

```python
# Menggabungkan DataFrame berdasarkan kolom 'Order Number'
merged_df = pd.merge(paid_transaction_df, seller_df, on='Order Number')
```

```python
# Buat kolom baru "Full Name" dengan format title case
merged_df["Full Name"] = merged_df["First Name"].str.title() + " " + merged_df["Last Name"].str.title()
```

```python
# Sisipkan kolom "Full Name" setelah kolom "Last Name"
columns = list(merged_df.columns)
last_name_index = columns.index("Last Name")
columns.insert(last_name_index + 1, "Full Name")
merged_df = merged_df.reindex(columns=columns)
```

```python
# Tampilkan DataFrame hasil
merged_df.head()
```

2. Menambahkan Kolom GMV

```python
merged_df["GMV"] = merged_df["Transaction Amount"] - merged_df["Seller Discount"] + merged_df["Delivery Fee"]
```

```python
# Sisipkan kolom "GMV" setelah kolom "Seller Discount"
columns = list(merged_df.columns)
seller_discount_index = columns.index("Seller Discount")
columns.insert(seller_discount_index + 1, "GMV")
merged_df = merged_df[columns]
```

```python
# Tampilkan DataFrame hasil
merged_df.head()
```

3. Mengelompokkan setiap Meta Category menjadi 2 group

```python
# Definisikan kategori grup
group_mapping = {
    'Home/Food': 'Group 1',
    'Sports/Hobi/Otomotif': 'Group 1',
    'Fashion': 'Group 1',
    'Beauty/Health': 'Group 1',
    'Gadget/Komputer': 'Group 2',
    'Elektronik': 'Group 2',
    'Service/Mokado': 'Group 2',
    'Babies/Kids': 'Group 2'
}
```

```python
merged_df["Group"] = merged_df["Meta Category"].map(group_mapping)
```

```python
# Create a pivot table with total GMV per month for each group
merged_df['Paid Date'] = pd.to_datetime(merged_df['Paid Date'])
merged_df['Month'] = merged_df['Paid Date'].dt.to_period('M')
```

```python
pivot_table = merged_df.pivot_table(
    values='GMV',
    index='Month',
    columns='Group',
    aggfunc='sum',
    fill_value=0
).reset_index()
```

4. Membuat python statement untuk mencari Seller yang memiliki total nilai GMV paling tinggi di bulan Agustus 2017

```python
df_august_2017 = merged_df[merged_df['Month'] == '2017-08']
top_seller_august = df_august_2017.groupby('Seller')['GMV'].sum().idxmax()
top_seller_august
```

5. Membuat python statement untuk mencari Seller yang memiliki paling banyak transaksi di bulan September 2017

```python
df_september_2017 = merged_df[(merged_df['Month'] == '2017-09') & (merged_df['Meta Category'] == 'Fashion')]
top_seller_september = df_september_2017['Seller'].value_counts().idxmax()
top_seller_september
```
